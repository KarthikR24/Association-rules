```python
# Import Data from CSV file 
import numpy as np
import pandas as pd
dataset = pd.read_csv('book.csv',sep=',')
dataset = dataset.replace(np.nan, '', regex=True)
dataset 
```

```python
dataset
transactions = []
for i in range(0, 2000):
    transactions.append([str(dataset.values[i,u]) for u in range(1, 10)]) 
```

```python
book_series  = pd.DataFrame(pd.Series(transactions))
book_series 
book_series.columns = ["business"]
book_series


```

```python
from mlxtend.frequent_patterns import apriori,association_rules
frequent_itemsets = apriori(dataset, min_support=0.005, max_len=3,use_colnames = True)
frequent_itemsets 
```

```python
frequent_itemsets.sort_values('support',ascending = False,inplace=True)
frequent_itemsets.sort_values
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head(20)
rules.sort_values('lift',ascending = False).head(10)
```
