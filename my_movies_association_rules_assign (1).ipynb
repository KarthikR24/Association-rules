```python
# Import Data from CSV file 
import numpy as np
import pandas as pd
dataset = pd.read_csv('my_movies.csv',sep=',')
dataset = dataset.replace(np.nan, '', regex=True)
dataset 
```

```python
dataset
transactions = []
for i in range(0, 10):
    transactions.append([str(dataset.values[i,u]) for u in range(1, 15)]) 
```

```python
my_movies_series  = pd.DataFrame(pd.Series(transactions))
my_movies_series 
my_movies_series.columns = ["business"]
my_movies_series
```

```python
X = my_movies_series['business'].str.join(sep=',').str.get_dummies(sep=',')
X
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
frequent_itemsets = apriori(X, min_support=0.005, max_len=3,use_colnames = True)
frequent_itemsets 
```

```python
frequent_itemsets.sort_values('support',ascending = False,inplace=True)
frequent_itemsets.sort_values
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head(20)
rules.sort_values('lift',ascending = False).head(10)
```
